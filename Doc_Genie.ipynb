```python
import ast
import textwrap

def read_python_file(file_path):
    with open(file_path, "r") as file:
        return file.read()
```

Function extract using AST

```python
def extract_functions(code):
    tree = ast.parse(code)
    functions = []

    for node in ast.walk(tree):
        if isinstance(node, ast.FunctionDef):
            functions.append(node)

    return functions
```

Docstring Generate Function

```python
def generate_docstring(function_node):

    func_name = function_node.name
    args = [arg.arg for arg in function_node.args.args]

    docstring = f"""
    Function: {func_name}

    Description:
    This function performs the operation of {func_name}.

    Parameters:
    """

    for arg in args:
        docstring += f"\n    {arg} : description"

    docstring += "\n\n    Returns:\n    description"

    return textwrap.dedent(docstring)
```

Display Docstrings

```python
def create_docstrings(code):

    functions = extract_functions(code)

    for func in functions:
        print(generate_docstring(func))
        print("-" * 50)
```
